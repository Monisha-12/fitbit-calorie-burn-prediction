1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

2. Load Dataset

```python
df = pd.read_csv("../data/raw/Fitbit_dataset.csv")

# Clean column names
df.columns = df.columns.str.strip()

# Drop unwanted column
if "Unnamed: 0" in df.columns:
    df = df.drop("Unnamed: 0", axis=1)

df.head()
```

3. Dataset Overview

```python
# The dataset contains 14102 rows and 18 columns including demographic, physiological, and workout-related features.
df.shape
print("Shape:", df.shape)
df.info() 
df.describe()
```

4. Missing Values Analysis

```python
df.isnull().sum()
# There are no / few missing values. Missing values were handled using median (numerical) and mode (categorical).
```

5. Feature Distributions

```python
# Numerical features
num_cols = df.select_dtypes(include=np.number).columns

df[num_cols].hist(figsize=(15,10), bins=20)
plt.tight_layout()
plt.show()
```

```python
# Categorical features
cat_cols = df.select_dtypes(include="object").columns

for col in cat_cols:
    plt.figure()
    sns.countplot(data=df, x=col)
    plt.title(col)
    plt.xticks(rotation=45)
    plt.show()

    # Most features are normally distributed, while some show skewness. Workout type and experience level show class imbalance.
```

6. Correlation Heatmap

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Strong correlation observed between heart rate features and calories burned. Derived features show very high correlation.
```

7. Leakage Detection

```python
# Step 1: Check correlation with target
target_col = "Calories_Burned (kcal)"

corr_with_target = df.corr(numeric_only=True)[target_col].sort_values(ascending=False)
print(corr_with_target)
```

```python
# Step 2: Visualize top correlated features
corr_with_target.plot(kind="bar", figsize=(10,5))
plt.title("Correlation with Target (Calories Burned)")
plt.show()
```

```python
# Distribution Plots
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()

    # Data Leakage Detection:
# Features such as Effective_MET, Base_MET, and HR_Intensity show extremely high correlation with the target variable.
# These features are derived from physiological calculations closely related to calorie burn.
# Including them in model training leads to artificially high accuracy and poor generalization.

# Therefore, these features will be removed before training the final model.
```

8. Remove Leakage Features (Preview)

```python
leakage_features = ["Base_MET", "Effective_MET", "HR_Intensity"]

df_clean = df.drop(columns=leakage_features)

df_clean.head()
```

9. Save Clean Dataset

```python
df_clean.to_csv("../data/processed/fitbit_clean.csv", index=False)

# Key Insights:

# Heart rate and session duration are strong predictors of calorie burn
# Derived features introduced data leakage and were removed
# Dataset contains a mix of physiological and behavioral features
# Data is suitable for both regression and clustering tasks
```
